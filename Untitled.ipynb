```python
 import os
import shutil
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator  # Eksik import
```

```python
# Görsellerin bulunduğu dizinler
rise_dir = '/Users/hasantekbas/Desktop/Trade/Up'  # Yükselme grafikleri
fall_dir = '/Users/hasantekbas/Desktop/Trade/Down'  # Düşüş grafikleri

# Yeni dizinler oluşturacağız
train_dir = '/Users/hasantekbas/Desktop/Trade/train_data'
test_dir = '/Users/hasantekbas/Desktop/Trade/test_data'
```

```python
# Eğer dizinler yoksa, onları oluştur
if not os.path.exists(train_dir):
    os.makedirs(train_dir)
if not os.path.exists(test_dir):
    os.makedirs(test_dir)
```

```python
# "rise" ve "fall" alt klasörlerini oluştur
train_rise_dir = os.path.join(train_dir, 'rise')
train_fall_dir = os.path.join(train_dir, 'fall')
test_rise_dir = os.path.join(test_dir, 'rise')
test_fall_dir = os.path.join(test_dir, 'fall')

# Eğer alt dizinler yoksa, onları oluştur
os.makedirs(train_rise_dir, exist_ok=True)
os.makedirs(train_fall_dir, exist_ok=True)
os.makedirs(test_rise_dir, exist_ok=True)
os.makedirs(test_fall_dir, exist_ok=True)
```

```python
# Yükselme ve Düşüş görsellerini toplayalım
rise_images = [os.path.join(rise_dir, fname) for fname in os.listdir(rise_dir)]
fall_images = [os.path.join(fall_dir, fname) for fname in os.listdir(fall_dir)]

# Görselleri eğitim ve test setlerine ayıralım
train_rise, test_rise = train_test_split(rise_images, test_size=0.2, random_state=42)  # %80 eğitim, %20 test
train_fall, test_fall = train_test_split(fall_images, test_size=0.2, random_state=42)
```

```python
# Eğitim verilerini ilgili dizine taşıyalım
for img in train_rise:
    shutil.copy(img, train_rise_dir)
for img in train_fall:
    shutil.copy(img, train_fall_dir)

# Test verilerini ilgili dizine taşıyalım
for img in test_rise:
    shutil.copy(img, test_rise_dir)
for img in test_fall:
    shutil.copy(img, test_fall_dir)
```

```python
# Verileri yüklemek için ImageDataGenerator kullanabiliriz
train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary'
)
```

```python
from tensorflow.keras import layers, models

# Modeli oluşturma
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    layers.MaxPooling2D((2, 2)),
    
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(1, activation='sigmoid')  # İki sınıf olduğu için sigmoid
])

# Modeli derleme
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# Modeli eğitme
history = model.fit(
    train_generator,
    steps_per_epoch=100,  # Eğitim verisinin toplam sayısına göre ayarlayın
    epochs=10,  # Eğitim sayısı (isteğe göre artırabilirsiniz)
    validation_data=test_generator,
    validation_steps=50  # Test verisinin toplam sayısına göre ayarlayın
)
```

```python
import matplotlib.pyplot as plt

# Eğitim ve doğrulama doğruluğu grafiği
plt.plot(history.history['accuracy'], label='Eğitim Doğruluğu')
plt.plot(history.history['val_accuracy'], label='Doğrulama Doğruluğu')
plt.title('Model Doğruluğu')
plt.xlabel('Epoch')
plt.ylabel('Doğruluk')
plt.legend(loc='lower right')
plt.show()

# Eğitim ve doğrulama kaybı grafiği
plt.plot(history.history['loss'], label='Eğitim Kaybı')
plt.plot(history.history['val_loss'], label='Doğrulama Kaybı')
plt.title('Model Kaybı')
plt.xlabel('Epoch')
plt.ylabel('Kaybı')
plt.legend(loc='upper right')
plt.show()
```

```python
# Modeli test verileriyle değerlendirme
test_loss, test_acc = model.evaluate(test_generator, steps=50)  # Test verisinin toplam sayısına göre ayarlayın
print(f'Test Kaybı: {test_loss}')
print(f'Test Doğruluğu: {test_acc}')
```

```python
from tensorflow.keras.preprocessing import image
import numpy as np

# Görseli yükleyip ön işleme
img_path = '/Users/hasantekbas/Desktop/Trade/test.png'  # Tahmin yapmak istediğiniz görselin yolu

# Görseli yükle
img = image.load_img(img_path, target_size=(150, 150))  # Modelin giriş boyutuna uygun şekilde

# Görseli numpy dizisine dönüştür
img_array = image.img_to_array(img)

# Boyutunu (1, 150, 150, 3) haline getirelim (batch size 1 için)
img_array = np.expand_dims(img_array, axis=0)

# Normalleştirme (Eğitimde kullanılan ile aynı olmalı)
img_array = img_array / 255.0
```

```python
# Model ile tahmin yapma
prediction = model.predict(img_array)

# Çıktıyı yorumlama
if prediction[0] > 0.5:
    print("Tahmin: Yükselme Grafiği")
else:
    print("Tahmin: Düşüş Grafiği")
```
